```python
import json
import pandas as pd
import requests
```

```python
def get_rewards_from_api (start_date, end_date, address):               
    URL = "https://api.helium.io/v1/hotspots/" + address + '/rewards/sum'
    response = requests.get(URL, params=[('min_time', start_date),('max_time', end_date)],)
    values = response.json()
    return values['data']['total']
```

```python
f = open('10-2021_hotspots.json')
data = json.load(f)


hotspots = {}
name = "hotspot_name"
add = "address"
for key,val in data.items():
    hotspots[val[name]] = val[add]
```

```python
print(len(hotspots))
```

```python
final = {}

for name,add in hotspots.items():
    temp = {}
    print('Current Hotspot: ',name)
    print('Address: ',add)
    for i in range(len(aug) - 1):
        start = aug[i]
        end = aug[i + 1]
        values = get_rewards_from_api(start, end, add)
        temp[start] = values
    final[name] = temp
```

```python
print(len(final))
```

```python
df = pd.DataFrame(final)
finaldf = df.transpose() 
print(finaldf)

finaldf.to_csv('Rewards_Aug21.csv',index=True)
```

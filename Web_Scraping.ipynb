```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
import urllib
from urllib.request import urlopen, Request
```

```python
url = "https://coinmarketcap.com/currencies/helium/historical-data/"
page = urllib.request.urlopen(url)
html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")
```

```python
print(soup.title.text)
```

```python
print(soup.prettify())
```

```python
tb = soup.find("div", { "class" : "sc-16r8icm-0 jKrmxw container" })
print(tb)

tb2 = soup.find("div", { "class" : "sc-114s7xe-0 ceRDaK" })
print(tb2)
```

```python
div = soup.find('table', {'class':'h7vnx2-2.jNaLNi.cmc-table'})
print(div)
```

```python
df = pd.read_html(str(div))[0]
print(df.head())
```
